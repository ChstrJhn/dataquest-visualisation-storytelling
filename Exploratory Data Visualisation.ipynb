## Findings so far

#### Do students in more popular majors make more money?

> No. Using 'Total' as a proxy for popularity of a major, there is no linear positive relationship between the total number of people with the major (business and social sciences being the most 'popular') and median salary. However, the opposite might be the case. 

#### Do students that majored in subjects that were majority female make more money?

> No. Again, the opposite seem to be the case. There seem to be a negative linear relationship between women dominated majors and median salary.

#### Is there any link between the number of full-time employees and median salary?

> There seem to be a negative linear relationship between the proportion of full-time employees and median salary. However, the upper quartile for full-time employees is dominated by majority business and social sciences majors, which can consist of a wide range of job types. With that said, looking at this relationship in particular might not bear any meaningful conclusions.
